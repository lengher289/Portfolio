# Automatic Monthly Attendance Data Extraction and Management Tool, TUTORIAL

## Prerequisites:

1. Before running the tool make sure that each individual participant has their own time sheet in their own respective folder.
2. Make sure that the file name that is in each folder is the same as well.

## Usage:

### Step 1: Install Dependencies
> To install the necessary Python libraries, run the following command by clicking on the play buton on the left:

## Step 3:Enter the File Name

> When prompted, type the name of the individual timesheet file, including the .docx extension (e.g., "9-2023 Participant Timesheet.docx"). If you're not sure, don't worry; it will remind you of the correct format.

enter file name.png

### After entering file name the program will ask a series of questions. Simply type your answer in the box and press enter.
